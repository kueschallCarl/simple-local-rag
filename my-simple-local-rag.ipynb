# Create and run local RAG Pipeline from Scratch

## What is RAG?

RAG stands for “Retrieval Augmented Generation”

The goal is to take information and pass it to an LLM so it can generate output based on that information

- Retrieval: Find relevant information given a query, e.g. "what are the macro nutrients and what do they do?" -> retrieves passages of text related to the macro nutrients from a Nutrition textbook
- Augmented: We want to take the relevant information and augment our input (prompt to LLM) with that relevant information.
- Generaton: Take the first two steps and pass them to an LLM for generative outputs.

## Why RAG
The main goal is to improve the generation outputs of LLMs, by providing additional relevant information that may be private in nature.

1. Prevent hallucinations: LLMs are good at generating good **looking** text, that doesn't mean it is factual however. RAG can help LLMs by providing relevant passages that are **factual**
2. Work with **custom data**. LLMs are trained on mostly internet data, which leads to a wide breadth of knowledge, that results in very generic responses for most cases.
    - Examples for custom data:
        - Lecture Notes
        - MyOTH website data

## What can RAG be used for more generally?
- Customer support Q&A chat: Treat your existing customer support documents as a resource and when a customer asks a question, use RAG to retrieve the relevant customer support document passages. LLM will use those snippets to craft a concise final answer.
- Email chain analysis: large insurance company that has lots of chains of customer emails. Use RAG to analyze these email chains -> The Pipeline can extract information from this data like SQL would from a database more or less
- Company internal documentation chat -> Confluence
- Textbook Q&A -> Like lecture notes

Common theme here: take your relevant documents and process them with LLM

From this angle, you can consider an LLM as a calculator for words
